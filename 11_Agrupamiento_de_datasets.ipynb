# Llamar los datos 

```
#Llamamos el set de datos público 
Chicago_data = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'

```

```
import pandas as pd 
```

```
Chicago_df=pd.read_csv(Chicago_data)
```

*Transformaciones * básicas  

```
Chicago_df['CRASH_DATE'] = Chicago_df['CRASH_DATE'].apply(lambda x: pd.to_datetime(x,errors='coerce', utc=True))
```

```
Chicago_df.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'BEAT_OF_OCCURRENCE':'sum',})
```

```
light_condition_per=Chicago_df.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'BEAT_OF_OCCURRENCE':'sum',})
```

```
light_condition_per=light_condition_per.reset_index()
```

```
light_condition_per.info()
```

```
import altair as alt 
```

```
Chicago_df.groupby(['LIGHTING_CONDITION','REPORT_TYPE']).agg({'CRASH_HOUR':'mean'})
```

```
Chicago_df.groupby(['LIGHTING_CONDITION','REPORT_TYPE']).agg({'CRASH_HOUR':['mean','min','max']})
```

```
crash_hours_reporte=Chicago_df.groupby(['LIGHTING_CONDITION','REPORT_TYPE']).agg({'CRASH_HOUR':['mean','min','max'],'CRASH_RECORD_ID':'count'})
```

```
crash_hours_reporte=crash_hours_reporte.reset_index()
```

```
crash_hours_reporte.head()
```

```
crash_hours_reporte.columns
```

```
new_column_names=['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR_mean','CRASH_HOUR_min','CRASH_HOUR_max','count_id']
```

```
 crash_hours_reporte.columns = new_column_names
```

```

alt.Chart(crash_hours_reporte).mark_bar().encode(
    x='LIGHTING_CONDITION',
    y='count_id',
    color='REPORT_TYPE'
).properties(width=220)
```
