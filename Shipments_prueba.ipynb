```
df_group = shipments_df.groupby("Labeling Carrier Name")
```

```
carrier_counts = shipments_df.groupby('Labeling Carrier Name')['Shipment ID'].count()
carrier_counts.plot(kind="bar",)
plt.grid()
plt.xlabel("Carriers Last Mile")
plt.ylabel("Q shipments")
plt.title("Cantidad de envíos por carrier")
```

```
status_counts = shipments_df.groupby('Hub Status')['Shipment ID'].count()
status_counts.plot(kind="bar",color="r")
plt.grid()
plt.xlabel("Hub status")
plt.ylabel("Q shipments")
plt.title("Cantidad de envíos por hub status")
```
